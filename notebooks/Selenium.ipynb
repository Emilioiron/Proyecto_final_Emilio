```python
data = browser.find_element_by_id('listaTrenesTable').text
# data = pd.DataFrame(browser.find_element_by_id('listaTrenesTable'))
data
```

```python
lista = list(data.split('\n'))   # create list split for \n
column_names = []   # create empty list
count = 0
for e in lista:
    count += 1
    if count <= 8:
        column_names.append(e)
column_names
```

```python
data = data.replace('Salida', '').strip('\n')
data = data.replace('Llegada', '').strip('\n')
data = data.replace('Duración', '').strip('\n')
data = data.replace('Servicio', '').strip('\n')
data = data.replace('Precio Desde', '').strip('\n')
data = data.replace('Clase', '').strip('\n')
data = data.replace('Tarifa', '').strip('\n')
data = data.replace('Opciones', '').strip('\n')
data = data.replace('Turista\nSólo plaza H\n', 'Turista_Sólo_plaza_H ') # ver lstrip y strip
data = data.replace('\nTren Completo', ' Tren Completo')
data = data.replace('\n\n', '\n')
data = data.replace('Tren Completo', 'None € Tren_Completo')
data = data.replace('Turista Plus', 'Turista_Plus')
data = data.replace('AVE','0 min. AVE')
data = data.replace('min. 0 min.','min.')
data
```
